```python
import json
from pathlib import Path
from collections import defaultdict

final_path = Path('marketeye_final.json')

with open(final_path, 'r', encoding='utf-8') as f:
    products = json.load(f)

print(f"üìä Total produits: {len(products)}")

# Trouver les produits avec plusieurs sources
multi_source_products = []

for product in products:
    offers = product.get('offers', [])
    sources = {offer.get('source') for offer in offers}
    
    if len(sources) > 1:
        multi_source_products.append({
            'id': product.get('product_id'),
            'sources': list(sources),
            'offers_count': len(offers),
            'prices': [offer.get('price') for offer in offers]
        })

print(f"üîç Produits sur plusieurs plateformes: {len(multi_source_products)}")

if multi_source_products:
    print("\nüìã Liste des produits multi-sources:")
    for p in multi_source_products[:10]:  # Limiter √† 10
        print(f"  {p['id']}: {p['sources']} ({p['offers_count']} offres)")
        for i, (source, price) in enumerate(zip(p['sources'], p['prices'])):
            print(f"    {source}: {price} MAD")
else:
    print("‚ÑπÔ∏è Aucun produit trouv√© sur plusieurs plateformes")
```
